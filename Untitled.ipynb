```python
df.head()
```

# Data Preprocessing

```python
counts=df['var_d7'].value_counts()
df=df[df['var_d7'].isin(counts[counts>100].index)]
```

```python
df.isnull().sum()
```

```python
df['var_d2']=df['var_d2'].fillna(df['var_d2'].mean())
df['var_d1']=df['var_d1'].fillna(df['var_d1'].dropna().mode().values[0])
df['var_d3']=df['var_d3'].fillna(df['var_d3'].dropna().mode().values[0])
df['var_d4']=df['var_d4'].fillna(df['var_d4'].dropna().mode().values[0])
df['var_d5']=df['var_d5'].fillna(df['var_d5'].dropna().mode().values[0])
df['var_d6']=df['var_d6'].fillna(df['var_d6'].dropna().mode().values[0])
df['var_d7']=df['var_d7'].fillna(df['var_d7'].dropna().mode().values[0])
df['var_d8']=df['var_d8'].fillna(df['var_d8'].dropna().mode().values[0])
```

```python
X=df.drop(['target'],axis=1)
X=pd.get_dummies(X)
X=X.drop(['var_d3_PUBLIC LTD COMPANIES', 'var_d6_ASSOCIATION', 'var_d7_AGENT', 'var_d7_CONSULTANT', 'var_d7_DOCTOR-SALARIED', 'var_d7_MEDIUM FLEETOPERATOR', 'var_d7_STUDENT'],axis=1)
y=df['target']
columns_train=X.columns
X.head()
X.shape
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler=RobustScaler()
X=robust_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
forest.fit(x_train,y_train)
```

```python
y_train_predict=forest.predict(x_train)
y_test_predict=forest.predict(x_test)
```
